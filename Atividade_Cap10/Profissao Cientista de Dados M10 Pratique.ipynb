```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

# Lendo o DataFrame

```python
# Definindo o caminho do arquivo e verificando se existe

caminho = r"C:\Users\MIGUEL\OneDrive\Programação\Primeiro Projeto\AtividadesEBAC\EBAC.DATA_SCIENCE\Atividade_Cap10\ecommerce_preparados.csv"

if os.path.exists(caminho):
    print("O caminho foi encontrado.")
else:
    print("O caminho não foi encontrado.")

# Lendo o arquivo CSV

df = pd.read_csv(caminho)

# Para ter uma ideia de que posso usar para faser os gráficos vou ver quais são as colunas da tabela com o segunte código:

print(df.columns)
```

# Criando os gráficos

```python
# Gráfico histograma
plt.figure(figsize=(12, 6))

plt.hist(df['Nota'], bins=28, color ='green',edgecolor='black') # Criando o gráfico 
plt.title('gráfico histograma - Notas x Numero de produtos com essa nota') 

plt.xlabel('Notas')
plt.xticks(ticks=np.arange(0, df['Nota'].max()+0.5, 0.5)) # Definindo os ticks de 0.5 para 0.5, aumento a exatidão de uma nota sem ficar massivo.

plt.ylabel('Quantidade')

plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersão

plt.figure(figsize = (12, 6))

plt.scatter(df['Desconto'], df['Qtd_Vendidos'], s=8, color='blue', alpha=0.5, edgecolors='black', linewidths=0.4) # Criando o gráfico
plt.title('Gráfico de dispersão - Porcentagem de desconto x Itens vendidos')

plt.xlabel('Desconto (%)')
plt.xticks(ticks=np.arange(0, df['Desconto'].max()+5, 5))

plt.ylabel('Quantidade de produtos vendidos')

plt.grid(True)
plt.xlim(0, df['Desconto'].max()+5) #Coloquei um limite para mostrar até onde ia os descontos 

plt.show()
```

```python
# Mapa de calor

cor = df[['Preço', 'Nota', 'N_Avaliações', 'Desconto']].corr() # Criando as correlações
sns.heatmap(cor, annot=True, cmap='coolwarm', linewidths=0.4, linecolor='black', cbar=True) # Criando o gráfico

plt.title('Mapa de calor - Preço x Notas x avaliações x Desconto')

plt.show()
```

```python
# Gráfico de barras 

plt.figure(figsize=(12, 6))

df['Gênero'].value_counts().plot(kind='bar', color='Red', edgecolor='black', linewidth=0.5) # Criando o gráfico

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.title('Gráfico de barras - Gênero x Quantidade')

plt.xlabel('Gênero')

plt.ylabel('Quantidade')

plt.show()
```

```python
# Gráfico de pizza 

temporadas_freq = df['Temporada'].value_counts().head(4) # Escolhi aqui pegar as 4 temporadas mais frequentes.

categorias = temporadas_freq.index # Pegando os nomes
frequencia = temporadas_freq.values # Pegando os valores

plt.figure(figsize=(12, 8))
plt.pie(frequencia, labels=categorias, autopct='%.1f%%', startangle=90, colors=['blue', 'red', 'green', 'orange']) # Criando o gráfico

plt.title('Gráfico de pizza - Temporada x Quantidade')

plt.show()
```

```python
# Gráfico de densidade

plt.figure(figsize=(12, 8))
sns.kdeplot(x=df['Desconto'], y=df['Nota'], fill=True, color='#863e9c') # Criando o gráfico, adicionei o fill para deixar mais bonito.

plt.title('Gráfico de densidade - Desconto x Nota')

plt.xlabel('Desconto (%)')

plt.ylabel('Nota')
plt.ylim(0, 5) # Coloquei um limite pra nota pois em teoria a nota máxima é 5.

plt.show()
```

```python
# Gráfico de Regressão 

sns.regplot(x='Desconto', y='Preço', data=df, color='red', scatter_kws={'alpha': 0.5, 'color': 'RoyalBlue'} )

plt.title('Gráfico de Regressão - Desconto x Preço')

plt.xlabel('Desconto (%)')

plt.ylabel('Preço')


plt.show()
```
