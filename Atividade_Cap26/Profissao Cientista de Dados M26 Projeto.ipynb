Abaixo temos a configuração do ambiente SQL:

```python
# Importando blibliotecas

import sqlite3
import pandas as pd
```

```python
# Criando os dfs

df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';', decimal=',')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
# Verificando se há valores nulos

print(df_transacoes.isnull().sum())
print(df_clientes.isnull().sum())
```

```python
# Fazendo as conexões para utilizar o sql pelo python

conn = sqlite3.connect('projeto.db')

# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

# Verificando as tabelas

```python
# Verificando tb_transacoes
query = """
SELECT DISTINCT
Id_client, Category, Price, Card_Type 
FROM TB_TRANSACOES
"""
result_df = run_query(query)
print(result_df)
```

```python
query = """
SELECT
*
FROM TB_CLIENTES
"""

result_df = run_query(query)
print(result_df)
```

```python
query = """
SELECT 
*
FROM TB_TRANSACOES AS A
LEFT JOIN TB_CLIENTES AS B
ON A.Id_client = B.Id_client
"""
df = run_query(query)
print(df)
```

Justifique a escolha do JOIN: 

Eu escolhi usar o left join, sendo tb_transacoes a tabela da esquerda. Tomei essa decisão pois considerei os dados de vendas e o que foi vendido os dados mais valiosos, e não queria perde-los pois mesmo que não seja possível identificar o cliente exato que comprou o valor da venda e o que foi vendido continua sendo muito importante para o negócio, logo, mante-los é essencial e a tb_transacoes é uma tabela maior, fazer right join ou inner, causaria uma perda de dados significativa o que está longe de ser ideal. Pensei em fazer full join, mas deixou muitos valores nulos, e os dados que foram acrescentados não são dados exatamentes confiáveis, pois se não é possivel encontrar a venda que o cliente fez, isso pode indicar que foi um cliente que fez o cadastro, mas não comprou nada, logo, utilizar essas métricas podem não trazer muitos benefícios, além de trazer mais valores nulos para o df.
